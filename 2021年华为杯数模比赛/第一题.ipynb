```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor 
from sklearn.model_selection import train_test_split 
from sklearn.datasets import load_boston
```

```python
#调用数据
data = pd.read_excel(r'Molecular_Descriptor.xlsx',header=0)
data1 = pd.read_excel(r'ERα_activity.xlsx',header=0)
```

```python
pIC50 = data1.values[:, -1]
```

```python
factors = data.drop(columns={'SMILES'})
factors = factors.dropna(axis=1) 
```

```python
feat_labels = data.columns[1:730]
```

```python

forest = RandomForestRegressor(n_estimators=500, 
                               random_state=0, 
                               max_features=100, 
                               n_jobs=2)
X_train,X_test,y_train,y_test=train_test_split(factors, pIC50,
                                    test_size=0.3,shuffle=True,random_state=0)
forest.fit(X_train,y_train)
importances = forest.feature_importances_ 
indices = np.argsort(importances)[::-1]
for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30, feat_labels[indices[f]], importances[indices[f]]))
```

```python
df = pd.DataFrame(importances)
```

```python
df.to_excel('importances.xlsx', index=False, header=False)
```
