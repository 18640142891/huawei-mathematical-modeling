```python
df = pd.read_excel(r'Molecular_Descriptor.xlsx',header =0)
data1 = pd.read_excel(r'ERα_activity.xlsx',header=0)
```

```python
factors = df.drop(columns={'SMILES'})
factors = factors.dropna(axis=1) 
```

```python
Y = pd.DataFrame(data=data1, columns=['pIC50'])
```

```python
def multipl(a,b):
    sumofab=0.0
    for i in range(len(a)):
        temp=a[i]*b[i]
        sumofab+=temp
    return sumofab
```

```python
def corrcoef(x,y):
    n=len(x)
    #求和
    sum1=sum(x)
    sum2=sum(y)
    #求乘积之和
    sumofxy=multipl(x,y)
    #求平方和
    sumofx2 = sum([pow(i,2) for i in x])
    sumofy2 = sum([pow(j,2) for j in y])
    num=sumofxy-(float(sum1)*float(sum2)/n)
    #计算皮尔逊相关系数
    den=sqrt((sumofx2-float(sum1**2)/n)*(sumofy2-float(sum2**2)/n))
    return num/den
```
